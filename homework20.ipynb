### frequency distribution

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
for word in tokenized_word:
    fdist[word.upper()]+=1
fdist    
```

```python
len(fdist)
```

```python
fdist_top20 = fdist.most_common(20)
fdist_top20
```

```python
from matplotlib.pyplot import figure
%matplotlib inline
import matplotlib.pyplot as plt


# Creating the word frequency distribution
freqdist = nltk.FreqDist(fdist_top20)
n_bins = 25


# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7),  tight_layout = True)
ax.hist(fdist_top20, bins = n_bins)
  
# Show plot
plt.show()
```

# Sentiment Analysis

```python
sents = gutenberg.sents('milton-paradise.txt')
print(len(sents))
```

```python
def to_list_sentences(sents):
    #takes in a list of lists
    sentence_list = []
    #for each list
    for sentence in sents:
        sentence_string =''
        # for each word in the list
        for word in sentence:
            sentence_string += ' ' + word
        temp = []
        temp.append(sentence_string)
        sentence_list.append(temp)
    return sentence_list
```

```python
sents_analysed=to_list_sentences(sents)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
# Generating sentiment for all the sentence present in the dataset
sentiments=[]
for sentence in sents_analysed:
    #print(sentence[0])
    sentiment = analyzer.polarity_scores(sentence[0])
    #print(sentiment)
    compound = sentiment["compound"]
    pos = sentiment["pos"]
    neu = sentiment["neu"]
    neg = sentiment["neg"]
    sentiments.append({
                    "text":sentence,
                    "compound":compound,
                    "positive":pos,
                    "negative":neg,
                    "neutral":neu
})
```

```python
import pandas as pd
#sentiments
df = pd.DataFrame(sentiments)
df
```
